```python
import numpy as np
from scipy.stats import wasserstein_distance
import seaborn as sns
```

```python
d1 = np.random.uniform(-1, 1, 1000)
d2 = np.random.normal(0, 2, 2000)
```

```python
sns.histplot(d1, bins=50, kde=False);
```

```python
sns.histplot(d2, bins = 50, kde = False);
```

```python
print(wasserstein_distance(d1, d2))
```
